```python
# pandasライブラリをインポート
import pandas as pd

# Ki_vs_Ta.csvファイルを読み込み、データフレームに変換する
df = pd.read_csv('./data/Ki_vs_Ta.csv')

# 上位3件の表示
print(df.head(3))
```

```python
#身長ソート(何もなし)
df = df.sort_values('身長')
df
```

```python
#体重ソート(降順)
df = df.sort_values(('体重'),ascending=False)
df
```

```python
#年代ソート(昇順) 
df = df.sort_values(('年代'),ascending=True)
df
```
